```python
def fy(x):
    return -12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30

def dy(x):
    return 12 * ùë• ** 4 * sin(ùë•) * cos(cos(ùë•)) - 48 * ùë• ** 3 * sin(cos(ùë•)) - 54 * ùë• ** 2 + 10 * ùë• + 10
```

```python
def bisection(f, lhs, rhs, eps=1e-10):
    mid = (lhs + rhs) / 2
    while abs(lhs - rhs) > eps:
        mid = (lhs + rhs) / 2
        if f(lhs) * f(mid) > 0: # f(lhs) > 0 and f(mid) > 0 or f(lhs) < 0 and f(mid) < 0
            lhs = mid
        else:
              rhs = mid      
    return mid
```
